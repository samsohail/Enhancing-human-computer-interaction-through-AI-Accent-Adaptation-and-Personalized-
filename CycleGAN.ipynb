```python
import torch
import torchaudio
from torch.utils.data import Dataset, DataLoader
import os
from torch.nn import functional as F
import torch.nn as nn
import torch.optim as optim
import librosa
import numpy as np
from scipy.io.wavfile import write
from audiomentations import Compose, AddGaussianNoise, TimeStretch, PitchShift, Shift
import matplotlib.pyplot as plt
import librosa.display

# Step 1: Define Data Augmentation Pipeline
augment = Compose([
    AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.015, p=0.5),
    TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5),
    PitchShift(min_semitones=-4, max_semitones=4, p=0.5),
    Shift(min_shift=-0.5, max_shift=0.5, p=0.5)
])

# Custom Dataset for Loading Audio Files with Augmentation
class SpeechDataset(Dataset):
    def __init__(self, data_dir, transform=None, target_length=80000, num_files=100, apply_augmentation=False):
        self.data_dir = data_dir
        self.transform = transform
        self.target_length = target_length
        self.apply_augmentation = apply_augmentation
        self.audio_files = [f for f in os.listdir(data_dir) if f.endswith('.wav')][:num_files]

        if len(self.audio_files) == 0:
            raise ValueError(f"No audio files found in directory: {data_dir}")

    def __len__(self):
        return len(self.audio_files)

    def __getitem__(self, idx):
        wav_path = os.path.join(self.data_dir, self.audio_files[idx])
        if not os.path.exists(wav_path):
            raise FileNotFoundError(f"Audio file not found: {wav_path}")

        try:
            waveform, sr = torchaudio.load(wav_path)
        except RuntimeError as e:
            print(f"Error loading audio file: {wav_path}, Error: {e}")
            raise e

        # Pad or truncate to the target length
        if waveform.shape[1] < self.target_length:
            padding = self.target_length - waveform.shape[1]
            waveform = F.pad(waveform, (0, padding))
        else:
            waveform = waveform[:, :self.target_length]

        # Apply Augmentation
        if self.apply_augmentation:
            waveform = augment_audio(waveform)

        # Apply normalization if any
        if self.transform:
            waveform = self.transform(waveform)

        return waveform, self.audio_files[idx]

# Function to Apply Augmentation to the Audio
def augment_audio(audio):
    augmented_samples = augment(samples=audio.numpy(), sample_rate=16000)
    return torch.tensor(augmented_samples)

# Define Normalization Transform
def normalize_waveform(waveform):
    return (waveform - waveform.mean()) / waveform.std()

# Directories for Data
data_dir_A = "/content/drive/MyDrive/data/extracted_files-3/en/North_American_English_W/"
data_dir_B = "/content/drive/MyDrive/data/extracted_files-3/en/British_English_W/"

# Target length for all audio files
target_length = 80000

# Initialize Datasets with 100 Files Each and Apply Augmentation
dataset_A = SpeechDataset(data_dir_A, transform=normalize_waveform, target_length=target_length, num_files=100, apply_augmentation=True)
dataset_B = SpeechDataset(data_dir_B, transform=normalize_waveform, target_length=target_length, num_files=100, apply_augmentation=True)

# Initialize DataLoaders
dataloader_A = DataLoader(dataset_A, batch_size=1, shuffle=True)
dataloader_B = DataLoader(dataset_B, batch_size=1, shuffle=True)

# Define Generators and Discriminators for CycleGAN
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(80000, 128),
            nn.ReLU(),
            nn.Linear(128, 80000)
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = self.main(x)
        x = x.view(x.size(0), 1, -1)
        return x

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(80000, 64),
            nn.LeakyReLU(0.2),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = x.view(x.size(0), -1)
        x = self.main(x)
        return x

# Initialize Models
G_AB = Generator()
G_BA = Generator()
D_A = Discriminator()
D_B = Discriminator()

# Initialize Loss Functions and Optimizers
criterion = nn.MSELoss()
optimizer_G = optim.Adam(list(G_AB.parameters()) + list(G_BA.parameters()), lr=0.0002, betas=(0.5, 0.999))
optimizer_D_A = optim.Adam(D_A.parameters(), lr=0.0002, betas=(0.5, 0.999))
optimizer_D_B = optim.Adam(D_B.parameters(), lr=0.0002, betas=(0.5, 0.999))

# Set device to GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

G_AB.to(device)
G_BA.to(device)
D_A.to(device)
D_B.to(device)

# Function to Compute MCD (Placeholder)
def compute_mcd(orig_audio, conv_audio):
    return np.random.random()

# Function to Plot Spectrograms
def plot_spectrogram(audio_path, title):
    y, sr = librosa.load(audio_path, sr=16000)
    plt.figure(figsize=(10, 4))
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()

# Evaluation function for Mean Mel-Cepstral Distortion (MCD)
def evaluate_metrics(original_dir, converted_dir, mapping):
    mcd_scores = []

    for orig_file, conv_file in mapping.items():
        orig_path = os.path.join(original_dir, orig_file)
        conv_path = os.path.join(converted_dir, conv_file)

        if not os.path.exists(orig_path):
            print(f"Original file not found: {orig_path}")
            continue
        if not os.path.exists(conv_path):
            print(f"Converted file not found: {conv_path}")
            continue

        try:
            orig_audio, _ = torchaudio.load(orig_path)
            conv_audio, _ = torchaudio.load(conv_path)
        except RuntimeError as e:
            print(f"Error loading audio files. Original: {orig_path}, Converted: {conv_path}, Error: {e}")
            continue

        # Compute MCD
        mcd_score = compute_mcd(orig_audio, conv_audio)
        mcd_scores.append(mcd_score)

        # Plot Spectrograms for the first audio pair (Original and Converted)
        if len(mcd_scores) == 1:
            plot_spectrogram(orig_path, f"Original: {orig_file}")
            plot_spectrogram(conv_path, f"Converted: {conv_file}")

    if mcd_scores:
        mean_mcd = np.mean(mcd_scores)
        print(f"Mean MCD: {mean_mcd:.4f}")
    else:
        print("No valid scores computed due to missing or corrupt files.")

# Function to save audio data correctly
def save_audio(file_path, audio_tensor, sample_rate=16000):
    audio_np = audio_tensor.cpu().detach().numpy().squeeze(0)
    audio_np = audio_np / np.max(np.abs(audio_np) + 1e-6)
    audio_np = np.clip(audio_np, -1, 1)
    audio_np = (audio_np * 32767).astype(np.int16)

    if len(audio_np.shape) > 1:
        audio_np = audio_np[0]

    write(file_path, sample_rate, audio_np)

# Training Loop
def train_cycleGAN(dataloader_A, dataloader_B, num_epochs=5, lambda_cycle=1.0):
    filename_mapping_A_to_B = {}
    filename_mapping_B_to_A = {}

    for epoch in range(num_epochs):
        batch_count = 0
        dataloader_A_iter = iter(dataloader_A)
        dataloader_B_iter = iter(dataloader_B)

        while True:
            try:
                (real_A, file_A), (real_B, file_B) = next(dataloader_A_iter), next(dataloader_B_iter)
            except StopIteration:
                break

            batch_count += 1
            print(f"Processing Epoch [{epoch+1}/{num_epochs}], Batch [{batch_count}]")

            real_A, real_B = real_A.to(device), real_B.to(device)
            file_A = file_A[0]
            file_B = file_B[0]

            # Train Discriminators
            optimizer_D_A.zero_grad()
            fake_A = G_BA(real_B).detach()
            loss_D_A = criterion(D_A(real_A), torch.ones_like(D_A(real_A)) * 0.9) + \
                       criterion(D_A(fake_A), torch.zeros_like(D_A(fake_A)) * 0.1)
            loss_D_A.backward()
            optimizer_D_A.step()

            optimizer_D_B.zero_grad()
            fake_B = G_AB(real_A).detach()
            loss_D_B = criterion(D_B(real_B), torch.ones_like(D_B(real_B)) * 0.9) + \
                       criterion(D_B(fake_B), torch.zeros_like(D_B(fake_B)) * 0.1)
            loss_D_B.backward()
            optimizer_D_B.step()

            # Train Generators
            optimizer_G.zero_grad()
            fake_B = G_AB(real_A)
            fake_A = G_BA(real_B)
            rec_A = G_BA(fake_B)
            rec_B = G_AB(fake_A)

            loss_G_adv = criterion(D_B(fake_B), torch.ones_like(D_B(fake_B))) + \
                         criterion(D_A(fake_A), torch.ones_like(D_A(fake_A)))
            loss_cycle = criterion(rec_A, real_A) + criterion(rec_B, real_B)
            loss_G = loss_G_adv + lambda_cycle * loss_cycle
            loss_G.backward()
            optimizer_G.step()

            # Save Converted Audio for Evaluation
            converted_A_to_B_path = f"/content/drive/MyDrive/data/extracted_files-3/en/converted-550_accentA_to_B_epoch_{epoch}/{file_A}"
            converted_B_to_A_path = f"/content/drive/MyDrive/data/extracted_files-3/en/converted-550_accentB_to_A_epoch_{epoch}/{file_B}"
            os.makedirs(os.path.dirname(converted_A_to_B_path), exist_ok=True)
            os.makedirs(os.path.dirname(converted_B_to_A_path), exist_ok=True)

            save_audio(converted_A_to_B_path, fake_B, sample_rate=16000)
            save_audio(converted_B_to_A_path, fake_A, sample_rate=16000)

            # Update Mappings for Evaluation
            filename_mapping_A_to_B[file_A] = file_A
            filename_mapping_B_to_A[file_B] = file_B

            print(f"Epoch [{epoch+1}/{num_epochs}], Batch [{batch_count}], Loss D_A: {loss_D_A.item():.4f}, Loss D_B: {loss_D_B.item():.4f}, Loss G: {loss_G.item():.4f}")

            if batch_count >= min(len(dataloader_A), len(dataloader_B)):
                break

        print(f"Epoch [{epoch+1}/{num_epochs}] completed with {batch_count} batches.")
        evaluate_cycleGAN(epoch, filename_mapping_A_to_B, filename_mapping_B_to_A)

    print("Training completed successfully!")

# Evaluation function to run after each epoch
def evaluate_cycleGAN(epoch, mapping_A_to_B, mapping_B_to_A):
    original_dir_A = "/content/drive/MyDrive/data/extracted_files-3/en/North_American_English_W/"
    converted_dir_A_to_B = f"/content/drive/MyDrive/data/extracted_files-3/en/converted-550_accentA_to_B_epoch_{epoch}"

    original_dir_B = "/content/drive/MyDrive/data/extracted_files-3/en/British_English_W/"
    converted_dir_B_to_A = f"/content/drive/MyDrive/data/extracted_files-3/en/converted-550_accentB_to_A_epoch_{epoch}"

    print(f"Evaluating CycleGAN performance after Epoch {epoch+1}")

    evaluate_metrics(original_dir=original_dir_A, converted_dir=converted_dir_A_to_B, mapping=mapping_A_to_B)
    evaluate_metrics(original_dir=original_dir_B, converted_dir=converted_dir_B_to_A, mapping=mapping_B_to_A)

# Start Training
train_cycleGAN(dataloader_A, dataloader_B)
```
